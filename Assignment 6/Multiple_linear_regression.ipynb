```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.feature_selection import RFE
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

##**Data Preperation**

```python
df = pd.read_csv('50_Startups.csv')
df.head()
```

```python
X = np.array(df.drop("Profit", axis="columns"))
Y = np.array(df["Profit"])
```

##**Data Pre-processing**

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

##**Multiple Linear Regression**

```python
LR = LinearRegression()
LR.fit(X_train, Y_train)
```

```python
Y_pred_lr = LR.predict(X_test)
Y_pred_lr = np.round(Y_pred_lr, decimals = 2)
```

```python
for i in range(len(Y_pred_lr)):
  print('predicted=%.2f, expected=%.2f' % (Y_pred_lr[i], Y[i]))
```

```python
print(LR.coef_)
print(LR.intercept_)
```

```python
print("RMSE = ", mean_squared_error(Y_test, Y_pred_lr, squared=False))
```

##**Selecting Best Features**

```python
rfe = RFE(LR, n_features_to_select=4)
rfe = rfe.fit(X_train, Y_train)
```

```python
print("Ranking of features: ", rfe.ranking_)
```

```python
Y_pred_rfe = rfe.predict(X_test)
Y_pred_rfe = np.round(Y_pred_rfe, decimals = 2)
```

```python
for i in range(len(Y_pred_rfe)):
  print('predicted=%.2f, expected=%.2f' % (Y_pred_rfe[i], Y[i]))
```

```python
print("RMSE = ", mean_squared_error(Y_test, Y_pred_rfe, squared=False))
```
