```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report
```

##**Data preparation**

```python
data = load_iris()
```

```python
X_raw = data.data
Y_raw = data.target
```

```python
X = X_raw[50:]
Y = Y_raw[50:] #Taking Iris-Versicolor and Iris-Virginica

X = X[:,1:3] #Taking sepal width and petal length
```

##**Data Analysis**

```python
scatter = plt.scatter(X[:, 0], X[:,1], c=Y)

plt.xlabel("Sepal width(cm)")
plt.ylabel("Petal length(cm)")

plt.legend(handles=scatter.legend_elements()[0], labels=["versicolor","virginica"], loc="upper left")
plt.tight_layout()
plt.show()
```

##**SVM**

```python
svm = SVC(kernel="linear")
svm.fit(X,Y)
```

```python
w = svm.coef_[0]
b = svm.intercept_[0]

step = (np.max(X[:,0]) - np.min(X[:,0]))/X.shape[0]
x_db = np.arange(np.min(X[:,0]), np.max(X[:,0]), step)

decision_boundary = -(b + w[0]*x_db)/w[1]
```

```python
SV1 = svm.support_vectors_
```

```python
scatter = plt.scatter(X[:, 0], X[:,1], c=Y)
plt.plot(x_db, decision_boundary, c="black", linestyle="--")
plt.scatter(SV1[:, 0], SV1[:,1], s=150, facecolors='none', edgecolors='black',  linestyle='--')

plt.xlabel("Sepal width(cm)")
plt.ylabel("Petal length(cm)")

plt.legend(handles=scatter.legend_elements()[0], labels=["versicolor","virginica"], loc="upper left")
plt.tight_layout()
plt.show()
```

##**Prediction**

```python
Y_pred = svm.predict(X)
```

```python
print(classification_report(Y, Y_pred))
```
