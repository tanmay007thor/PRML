```python
import numpy as np
```

```python
from sklearn.experimental import enable_iterative_imputer
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import normalize
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
```

```python
X1 = [[1, -1, 2],
      [2, 0, 0],
      [0, 1, -1]]

X2 = [['male', 'US', 'Safari'],
      ['female', 'Europe', 'Firefox'],
      ['female', 'Asia', 'Chrome']]

X3 = [['male'],
      ['female'],
      [np.nan],
      ['female']]

X4 =[['male', 'Safari'],
     ['female', None],
     [np.nan, 'Firefox']]

X5 = [[1, 2],
      [3, 6],
      [4, 8],
      [np.nan, 3],
      [7, np.nan]]

X6 = [["a", "x"],
      [np.nan, "y"],
      ["a", np.nan],
      ["b", "y"]]
```

##**Standard Scaler**

```python
scaler = StandardScaler()
X_new_ss = scaler.fit_transform(X1)
print(np.array(X1))
print('\n')
print(X_new_ss)
```

##**Min Max Scaler**

```python
min_max_scaler = MinMaxScaler()
X_new_mm = min_max_scaler.fit_transform(X1)
print(np.array(X1))
print('\n')
print(X_new_mm)
```

##**Normalization**

```python
X_new_l2 = normalize(X1, norm="l2")
print(np.array(X1))
print('\n')
print("l2 normalization")
print(X_new_l2)
```

```python
X_new_l1 = normalize(X1, norm="l1")
print(np.array(X1))
print('\n')
print("l1 normalization")
print(X_new_l1)
```

##**Ordinal Encoder**

```python
oe = OrdinalEncoder()
X_new1_oe = oe.fit_transform(X2)
print(np.array(X2))
print('\n')
print(X_new1_oe)
```

```python
oe = OrdinalEncoder()
X_new2_oe = oe.fit_transform(X3)
print(np.array(X3))
print('\n')
print(X_new2_oe)
```

```python
oe = OrdinalEncoder(encoded_missing_value=-1)
X_new3_oe = oe.fit_transform(X3)
print(np.array(X3))
print('\n')
print(X_new3_oe)
```

##**One Hot Encoder**

```python
ohe = OneHotEncoder()
X_new1_ohe = ohe.fit_transform(X2).toarray()
print(np.array(X2))
print('\n')
print(X_new1_ohe)
```

```python
ohe = OneHotEncoder(drop="first")
X_new2_ohe = ohe.fit_transform(X2).toarray()
print(np.array(X2))
print('\n')
print(X_new2_ohe)
```

```python
ohe = OneHotEncoder(handle_unknown="error")
X_new3_ohe = ohe.fit_transform(X4).toarray()
print(np.array(X4))
print('\n')
print(X_new3_ohe)
```

##**Univariate Feature Imputation**

```python
ufi = SimpleImputer(missing_values=np.nan, strategy="mean")
X_new1_ufi = ufi.fit_transform(X5)
print(np.array(X5))
print('\n')
print(X_new1_ufi)
```

```python
ufi = SimpleImputer(strategy="most_frequent")
X_new2_ufi = ufi.fit_transform(X6)
print(np.array(X6))
print('\n')
print(X_new2_ufi)
```

##**Multivariate Feature Imputation**

```python
mfi = IterativeImputer(max_iter=10, random_state=0)
X_new_mfi = np.round(mfi.fit_transform(X5))
print(np.array(X5))
print('\n')
print(X_new_mfi)
```
