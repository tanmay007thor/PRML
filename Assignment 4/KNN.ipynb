```python
sp_names = [0, 1]
```

##**K-NN**

```python
def diagram(X1,X2,Y,new_data,result):
  temp_x1 = np.append(X1, new_data[0])
  temp_x2 = np.append(X2, new_data[1])
  temp_y = np.append(Y, result)

  plt.scatter(temp_x1, temp_x2, c=temp_y)
  plt.scatter(new_data[0], new_data[1], s=150, facecolors='none', edgecolors='red', linestyle='--')

  plt.show()
```

```python
def knn(X1, X2, Y, k, new_data):
  ED1 = (X1-new_data[0])**2
  ED2 = (X2-new_data[1])**2

  ED = np.sqrt(ED1 + ED2)

  Y_sort = Y[np.argsort(ED)]

  result = 0
  for i in range(k):
    result += Y_sort[i]

  if(result > 1):
    diagram(X1,X2,Y,new_data,1)
  else:
    diagram(X1,X2,Y,new_data,0)
```

```python
k = 3
knn(X1, X2, Y, k, [25,40])
```
