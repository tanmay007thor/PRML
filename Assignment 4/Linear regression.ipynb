```python
import numpy as np
import matplotlib.pyplot as plt
```

#**data preparation**

```python
X = np.array([1, 2, 4, 5, 5, 6, 6, 7, 8, 10, 11, 11, 12, 12, 14]) #Total hours studied
Y = np.array([64, 66, 76, 73, 74, 81, 83, 82, 80, 88, 84, 82, 91, 93, 89]) #Final exam scores
```

```python
plt.plot(X, Y)
plt.xlabel("Total hours studied")
plt.ylabel("Final exam scores")
plt.show()
```

#**linear regression**

```python
n = X.shape[0]
```

```python
m = np.dot(X-np.mean(X), Y-np.mean(Y))/np.dot(X-np.mean(X), X-np.mean(X))
```

```python
b = (np.sum(Y) - m*np.sum(X))/n
```

```python
print(m)
print(b)
```

##**prediction**

```python
Y_pred = m*X + b
```

```python
plt.plot(X, Y, c="blue", label="Original")
plt.plot(X, Y_pred, c="red", label="Predicted")
plt.xlabel("Total hours studied")

plt.legend()
plt.show()
```
